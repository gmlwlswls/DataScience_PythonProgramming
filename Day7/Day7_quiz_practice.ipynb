```python
# 2) Reverse SLL

class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

# sol)
def created_linked_list(lst) : # append() 함수 방식 - cur, next 변수 필요
    if len(lst) == 0 :
        return None
    else :
        head_node = LinkedNode(lst[0])
        tail_node = head_node
        for i in range(1, len(lst)) :
            new_node = LinkedNode(lst[i])
            tail_node.next = new_node
            tail_node = new_node
        return head_node

def print_linked_list(head) :
    linked_list = []
    while head != None :
        linked_list.append(head.val)
        head = head.next
    print(linked_list)

def reverse_SLL(head) : # append() > addfirst 함수 방식 - prev, cur, next 변수 전부 필요
    cur_node = head
    prev_node = None # prev node를 통해 addfirst 구현
    while cur_node != None :
        next_node = cur_node.next
        cur_node.next = prev_node
        prev_node = cur_node # (1, None) > (2, (1, None)) 
        cur_node = next_node
    return prev_node
```

```python
l1 = created_linked_list([1,2,3,4,5,6,7])
l2 = created_linked_list([])

print_linked_list(l1)
print('/Expected: [1,2,3,4,5,6,7]')

print_linked_list(reverse_SLL(l1))
print('/Expected: [7,6,5,4,3,2,1]')

print_linked_list(reverse_SLL(l2))
print('/Expected: []')
```

```python
# 3-1) def reverseSLL 활용

class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

def reverse_SLL(head) : # append() > addfirst 함수 방식 - prev, cur, next 변수 전부 필요
    cur_node = head
    prev_node = None # prev node를 통해 addfirst 구현
    while cur_node != None :
        next_node = cur_node.next
        cur_node.next = prev_node
        prev_node = cur_node # (1, None) > (2, (1, None)) 
        cur_node = next_node
    return prev_node

def isPalindrome(head):
    reverse_head = reverse_SLL(head)
    while head != None :
        if head != reverse_head :
            return False
        else :
            head = head.next
            reverse_head = reverse_head.next
    return True
```
