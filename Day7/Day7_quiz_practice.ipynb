```python
# 2) Reverse SLL

class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

# sol)
def created_linked_list(lst) :
    if len(lst) == 0 :
        return None
    start = LinkedNode(lst[0])
    current_node = start
    for i in range(1, len(lst)) :
        linked_node = LinkedNode(lst[i])
        current_node.next = linked_node # (1, (2, None)) > (1, (2, (3, None)))
        current_node = linked_node # (2, None) > (3, None)
    return start
        
def print_linked_list(head):
    if head == None :
        return None
    else :
        linked_list = []
        linked_list.append(head.val)
        current_node = head
        while current_node.next != None :
            linked_list.append(current_node.next.val)
            current_node = current_node.next
        print('linked_list: ', linked_list)
    

"""
Implement function `reverse_SLL` that takes in the head of a SLL and returns the head of a reversed SLL

※ Space complexity should be O(1)  
(Generating new linked lists or lists is not allowed)
"""

def reverse_SLL(head):
    current_node = head
    prev_node = None
    while current_node != None :
        next_node = current_node.next
        current_node.next = prev_node
        prev_node = current_node
        current_node = next_node
    return prev_node
```
