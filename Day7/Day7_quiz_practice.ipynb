```python
# 1. myStack sol)

class LinkedNode() :
  def __init__(self, x):
    self.val = x
    self.next = None

class myStack :
  def __init__(self):
     self.sentinel = LinkedNode(0)
     self.size = 0
     
  def push(self, x) :
    self.size += 1
    newNode = LinkedNode(x) # Create a new LinkedNode object.
    newNode.next = self.sentinel # Connect it next to the sentinel.
    self.sentinel.next = newNode 
  
  def pop(self) :
    self.size -= 1
    self.sentinel.next = self.sentinel.next.next

  def top(self) :
    if self.sentinel.next == None :
      return None
    return self.sentinel.next.val
  
  def getSize(self) :
    return self.size
  
  def isEmpty(self) :
    return self.size == 0
  
  def clear(self) :
    self.sentinel.next = None # Make size 0 and remove sentinel's connection with the next node
    self.size = 0

    # Do not modify
  def status_check(self):
      if self.isEmpty():
          print("IsEmpty:",self.isEmpty(), " | Size:", self.getSize(),"| Top:", self.top(), end = " ")
      else:
          print("IsEmpty:",self.isEmpty(), "| Size:", self.getSize(),"| Top:", self.top(),  end = " ")
```

```python
# 2) Reverse SLL

class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

# sol)
def created_linked_list(lst) : # append() 함수 방식 - cur, next 변수 필요
    if len(lst) == 0 :
        return None
    else :
        head_node = LinkedNode(lst[0])
        tail_node = head_node
        for i in range(1, len(lst)) :
            new_node = LinkedNode(lst[i])
            tail_node.next = new_node
            tail_node = new_node
        return head_node

def print_linked_list(head) :
    linked_list = []
    while head.next != None : # head.next 기준으로 (7, None) 미포함
        linked_list.append(head.val)
        head = head.next
    linked_list.append(head.val)
    print(linked_list)

def reverse_SLL(head) : # append() > addfirst 함수 방식 - prev, cur, next 변수 전부 필요
    cur_node = head
    prev_node = None # prev node를 통해 addfirst 구현
    while cur_node != None :
        next_node = cur_node.next
        cur_node.next = prev_node
        prev_node = cur_node # (1, None) > (2, (1, None)) 
        cur_node = next_node
    return prev_node
```

```python
# 7-3-1) def reverseSLL 활용 sol
 
 # Return in boolean
 # Try doing it without reversing the whole linked list!(reverse_SLL func)
 
class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

def reverse_SLL(head) : # append() > addfirst 함수 방식 - prev, cur, next 변수 전부 필요
    cur_node = head
    prev_node = None # prev node를 통해 addfirst 구현
    while cur_node != None :
        next_node = cur_node.next
        cur_node.next = prev_node
        prev_node = cur_node # (1, None) > (2, (1, None)) 
        cur_node = next_node
    return prev_node

def isPalindrome(head):
    reverse_head = reverse_SLL(head)
    while head != None : # <=> while head
        if head != reverse_head :
            return False
        else :
            head = head.next
            reverse_head = reverse_head.next
    return True
```

```python
# 7-3-2 sol) reverseSLL_활용x
# ** You Only Need to reverse half of the linked list!

class LinkedNode():
    def __init__(self,x):
        self.val = x
        self.next = None

def isPalindrome(head) :
  reverse = None
  fast = slow = head 
  # slow보다 fast가 2배 빠르게 움직임
  # fast는 2칸씩 slow는 1칸씩
  while fast and fast.next :
  # 두 칸씩 이동하니까 fast와 fast.next 두 개를 None이 아닌지 확인
    fast = fast.next.next
    reverse, reverse.next, slow = slow, reverse, slow.next # slow가 reverse의 next역할

  if fast : # 만약 fast.next에서 None이었던거라면(홀수라서)
    slow = slow.next
  # 여기까지 하면 slow는 절반에 가있음
  while slow :
    if slow.val != reverse.val :
      return False
    else :
      slow = slow.next
      reverse = reverse.next
  return True
```
